### 1.1. <a name="sect11"></a>Test if
La forme complète d'un test utilise les mots-clés suivant : if, elif et else.

Vous pouvez mettre autant de elif que vous voulez après une condition en if. 
Tout comme le else, cette instruction est facultative et, quand bien même vous construiriez une instruction en if,elif, vous n'êtes pas du tout obligé de prévoir un else après. 
En revanche, l'instruction else ne peut figurer qu'une fois, clôturant le bloc de la condition.

**Exercice** : Déterminer si un mot saisi par l'utilisateur contient la lettre 'e'

Si le mot contient la lettre 'e', afficher "La lettre e est présente dans ce mot". Si non, afficher "La lettre e n'est pas présente dans ce mot"

A noter que la demande d'une valeur à l'utilisateur se fait par la fonction **input**.

```python
mot=input("Donne un mot: ")

if "e" in mot:
    print('ok')
    index=mot.find("e")
    print(f"La lettre 'e' est présente dans ce mot à la {index +1}ème lettre")
else:
    print("La lettre e n'est pas présente dans ce mot")
```

### <a name="sect12"></a>1.2. Boucle while
Elle permet de réexécuter un bloc d'instruction tant qu'une condition est vrai.

Refaire l'exercice précédent en vérifiant que l'utilisateur a bien saisi une année (i.e. un nombre). Si ce n'est pas le cas, demander lui de recommencer jusqu'à ce que cela le soit.

```python
annee = input("Saisissez une année : ")
while not annee.isnumeric() :
    annee = input("Ceci n'est pas un nombre. Saisissez une année : ")
    
annee = int(annee)
if annee % 400 == 0 or (annee % 4 == 0 and annee % 100 != 0):
    print("L'année saisie est bissextile.")
else:
    print("L'année saisie n'est pas bissextile.")
```

### 2.2. <a name="sect22"></a>Def

**Exercice** : créer une fonction qui prend en paramètre une chaine de caractère et qui renvoie un dictionnaire avec
 - chaque mot comme clé
 - un boolean indiquant si c'est un palindrome comme valeur
 
Pour rappel, un palindrome est un mot qui est symétrique (ex: kayak, bob, radar, stats, coloc, ...)

```python
# Version 1
def chaine(chaine):
    dico={}
    for mot in chaine.lower().split():
        for i in range(len(mot)//2):
            if mot[i] == mot[-i-1]:
                bol="True"
            else: 
                bol="False"
        dico[mot]=bol
    return(dico)
```

```python
chaine("Python est mieux que Sas")
```

```python
# Version 2
def est_palindrome(chaine):
    res = {}
    for mot in chaine.lower().split():
        if mot == mot[::-1]:
            res[mot] = True
        else : res[mot]=False
    return res
```

Puis l'appliquer sur la chaine suivante : "Python est mieux que SAS"

```python
est_palindrome("Python est mieux que Sas")
```

**Etape 2** : Créer une fonction "standardisation" qui va centrer et réduite votre vecteur.

Pour rappel,
 - centrer = prendre chaque valeur et lui retirer la moyenne du vecteur
 - réduire = prendre chaque valeur et diviser par l'écart-type


```python
def standardisation(x): 
    x -= np.mean(x) # the -= means can be read as x = x- np.mean(x)
    x /= np.std(x)
    return x
```

**Etape 3** : modifier la fonction "standardisation" pour quelle intègre la gestion de cette erreur

```python
def standardisation(x): 
    try:        
        x-=np.mean(x) # the -= means can be read as x = x- np.mean(x)
        x/=np.std(x)
        return x
    except TypeError:
        print("le type des données n'est pas float")
```

```python
print(standardisation(vecteur1))
standardisation(vecteur2)
```

## BONUS <a name="sectBonus"></a>
L'objectif est de créer une fonction "infos" qui prend en paramètre une liste et qui renvoie les informations suivantes sous forme de dictionnaire :
 - si la liste contient des valeurs numériques :
     - Nombre de valeurs
     - Moyenne
     - Ecart-type
     - Minimum
     - 1er quartile
     - Mediane
     - 3ème quartile
     - Maximum
 - si la liste contient des chaines de caractères :
     - Nombre de valeurs
     - Nombre de valeurs distinctes
     - Valeur la plus fréquente
     - Nombre de fois où elle est présente

```python
from statistics import mean, stdev, median, mode
import numpy as np
from collections import Counter
```

```python
def infos(liste):
    res={}
    if all(isinstance(x, int) for x in liste) :
        res['nb_valeur'] = len(liste)
        res['moyenne'] = mean(liste)
        res['ecart_type'] = stdev(liste)
        res['minimum'] = min(liste)
        res['Q1'] = np.percentile(liste, 25)
        res['mediane'] = median(liste) # ou np.percentile(liste, 50)
        res['Q3'] = np.percentile(liste, 75)
        res['maximum'] = max(liste)
    elif all(isinstance(x, str) for x in liste) :
        res['nb_valeurs'] = len(liste)
        res['nb_unique_valeurs'] = len(np.unique(liste)) # ou len(set(liste)) # ou len(Counter(liste).most_common())
        res['top'] = mode(liste) # ou Counter(liste).most_common(1)[0][0]
        res['frequence'] = Counter(liste).most_common(1)[0][1]
    else :
        res['erreur'] = "La liste ne contient pas des éléments de même type"
    return res
```

Puis la tester sur les listes suivantes :

```python
liste1 = [1,2,4,5,6,8,9,13,42]
liste2 = ["bonjour","hello","holà","bonjour","hallo", "nǐ hăo","aloha","buongiorno","bonjour","nǐ hăo"]
liste3 = [1,2,4,5,6,8,9,13,42,"bonjour","hello","holà","bonjour","hallo", "nǐ hăo","aloha","buongiorno","bonjour","nǐ hăo"]
```

```python
infos(liste1)
```

```python
infos(liste2)
```
